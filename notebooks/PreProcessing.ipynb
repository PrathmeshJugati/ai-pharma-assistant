### Preprocessing

```python
dt.info()
    
```

```python
dt.isnull().sum()
```

```python
dt['composition'] = dt['short_composition1'].str.cat(dt['short_composition2'], na_rep="")
dt['composition'] = dt['composition'].str.lower()
new_dt = dt.drop(['short_composition1','short_composition2','salt_composition'], axis = 1)
```

```python
new_dt = new_dt.drop_duplicates(subset=["name", "manufacturer_name", "composition"])
new_dt = new_dt[new_dt["Is_discontinued"] == 0]
```

```python
new_dt['cleaned_name'] = new_dt['name'].str.lower().str.strip()
new_dt['pack_size_label'] = new_dt['pack_size_label'].str.lower().str.strip()
new_dt['manufacturer_name'] = new_dt['manufacturer_name'].str.lower().str.strip()
```

```python
new_dt['price'] = new_dt['price'].astype(int)
```

```python
new_dt.head(5)
```

```python
def create_doc(row):
    return f"""
    Medicine : {row['cleaned_name']},
    Composition : {row['composition']},
    Manufacturer : {row['manufacturer_name']},
    Type : {row['type']},
    Pack Size : {row['pack_size_label']},
    Price : {row['price']} INR
    """
```

```python
new_dt['document'] = new_dt.apply(create_doc, axis = 1)
```
